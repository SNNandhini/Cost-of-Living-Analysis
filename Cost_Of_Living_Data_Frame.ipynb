```python
#Import dependencies

import os
import csv
import pandas as pd
```

```python
# Create FilePath for CSV file

csv_path = "Data/Cost_of_living_indices_data.csv"
df = pd.read_csv(csv_path)

# Output
df
```

```python
# Create data for top 10 performing countries based on Cost of Living Index

Df_1 = df.sort_values(["Cost of Living Index"], ascending=False)


# Top 10 Countries DataFrame Output

top_10_countries_df = Df_1.head(10)

# Output
top_10_countries_df
```

```python
# Save Top 10 Countries as CSV file
top_10_countries_df.to_csv("Top_10_Countries.csv", index=False)
```

```python
# Bottom 10 countries Output

Bottom_10_Countries_df = Df_1.tail(10)

Bottom_10_Countries_df
```

```python
# Save Bottom 10 countries as CSV 
Bottom_10_Countries_df.to_csv("Bottom_10_Countries.csv", index=False)
```
